```python
from tabulate import tabulate
def matrix_chain_order_bottom_up(p):
    n = len(p) - 1
    m = [[0] * n for _ in range(n)]
    s = [[0] * n for _ in range(n)]
    for l in range(2, n + 1):
        for i in range(n - l + 1):
            j = i + l - 1
            m[i][j] = float('inf')
            for k in range(i, j):
                q = m[i][k] + m[k + 1][j] + (p[i] * p[k + 1] * p[j + 1])
                if q < m[i][j]:
                    m[i][j] = q
                    s[i][j] = k
    print("M Matrix:")
    print(tabulate(m[::-1], tablefmt='grid'))  
    print("\nS Matrix:")
    print(tabulate(s[::-1], tablefmt='grid')) 
    print("\nOptimal Parenthesization:", end=" ")
    print_optimal_parenthesis(s, 0, n - 1)
def print_optimal_parenthesis(s, i, j):
    if i == j:
        print(f"A{i + 1}", end="")
    else:
        print("(", end="")
        print_optimal_parenthesis(s, i, s[i][j])
        print_optimal_parenthesis(s, s[i][j] + 1, j)
        print(")", end="")
def main_bottom_up():
    p = []
    m = int(input("Enter the number of matrices: "))
    for i in range(m + 1):
        j = int(input("Enter the dimension of matrix " + str(i + 1) + ": "))
        p.append(j)
    matrix_chain_order_bottom_up(p)
main_bottom_up()
```

```python
from tabulate import tabulate
def matrix_chain_order_top_down(p):
    n = len(p) - 1
    m = [[float('inf')] * n for _ in range(n)]
    s = [[0] * n for _ in range(n)]
    lookup_chain(m, s, p, 0, n - 1)
    
    print("M Matrix:")
    print(tabulate(m[::-1], tablefmt='grid'))  
    print("\nS Matrix:")
    print(tabulate(s[::-1], tablefmt='grid'))  
    print("\nOptimal Parenthesization:", end=" ")
    print_optimal_parenthesis_top_down(s, 0, n - 1)
def lookup_chain(m, s, p, i, j):
    if m[i][j] < float('inf'):
        return m[i][j]
    if i == j:
        m[i][j] = 0
    else:
        for k in range(i, j):
            q = lookup_chain(m, s, p, i, k) + lookup_chain(m, s, p, k + 1, j) + (p[i] * p[k + 1] * p[j + 1])
            if q < m[i][j]:
                m[i][j] = q
                s[i][j] = k
    return m[i][j]
def print_optimal_parenthesis_top_down(s, i, j):
    if i == j:
        print(f"A{i + 1}", end="")
    else:
        print("(", end="")
        print_optimal_parenthesis_top_down(s, i, s[i][j])
        print_optimal_parenthesis_top_down(s, s[i][j] + 1, j)
        print(")", end="")
def main_top_down():
    p = []
    m = int(input("Enter the number of matrices: "))
    for i in range(m + 1):
        j = int(input("Enter the dimension of matrix " + str(i + 1) + ": "))
        p.append(j)
    matrix_chain_order_top_down(p)
main_top_down()
```

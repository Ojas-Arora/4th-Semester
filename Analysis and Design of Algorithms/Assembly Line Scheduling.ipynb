```python
from tabulate import tabulate
def assembly_line(station, t, e1, e2, x1, x2):
    n = len(station[0])
    f = [[0] * n for _ in range(2)]
    decision = [[0] * n for _ in range(2)]
    f[0][0] = e1 + station[0][0]
    f[1][0] = e2 + station[1][0]
    for j in range(1, n):
        if f[0][j - 1] + station[0][j] <= f[1][j - 1] + t[1][j - 1] + station[0][j]:
            f[0][j] = f[0][j - 1] + station[0][j]
            decision[0][j] = 1
        else:
            f[0][j] = f[1][j - 1] + t[1][j - 1] + station[0][j]
            decision[0][j] = 2
        if f[1][j - 1] + station[1][j] <= f[0][j - 1] + t[0][j - 1] + station[1][j]:
            f[1][j] = f[1][j - 1] + station[1][j]
            decision[1][j] = 2
        else:
            f[1][j] = f[0][j - 1] + t[0][j - 1] + station[1][j]
            decision[1][j] = 1
    print(tabulate(f, tablefmt="grid"))
    print(tabulate(decision, tablefmt="grid"))
    if f[0][n - 1] + x1 <= f[1][n - 1] + x2:
        final = f[0][n - 1] + x1
        l = 1
    else:
        final = f[1][n - 1] + x2
        l = 2
    print("\n\nSelected Lines at each station: ")
    print_stations(decision, l, n)
    return final, l
def print_stations(decision, l, n):
    i = l - 1
    print("Line " + str(i + 1) + " station " + str(n))
    for j in range(n - 1, 0, -1):
        i = decision[i][j]
        print("Line " + str(i) + " station " + str(j))
        i = i - 1
e1 = int(input("Enter the entry time of assembly line 1: "))
e2 = int(input("Enter the entry time of assembly line 2: "))
x1 = int(input("Enter the exit time of assembly line 1: "))
x2 = int(input("Enter the exit time of assembly line 2: "))
station1 = []
station2 = []
m = int(input("Enter the number of stations on one assembly line: "))
for i in range(m):
    j = int(input("Enter the amount of work at station " + str(i + 1) + " of assembly line 1: "))
    k = int(input("Enter the amount of work at station " + str(i + 1) + " of assembly line 2: "))
    station1.append(j)
    station2.append(k)
t1 = []
t2 = []
for i in range(m - 1):
    j = int(input("Enter the amount of work to change station from 1 to 2 at station " + str(i + 1) + " : "))
    k = int(input("Enter the amount of work to change station from 2 to 1 at station " + str(i + 1) + " : "))
    t1.append(j)
    t2.append(k)
station = [station1, station2]
t = [t1, t2]
min_cost, line = assembly_line(station, t, e1, e2, x1, x2)
print("\nMinimum cost is " + str(min_cost) + " from assembly line " + str(line))
```

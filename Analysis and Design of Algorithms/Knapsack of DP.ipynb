```python
from tabulate import tabulate

def knapsack_memoization(p, wt, m, n, memo):
    if memo[n][m] != -1:
        return memo[n][m]
    
    if n == 0 or m == 0:
        memo[n][m] = 0
    elif wt[n - 1] > m:
        memo[n][m] = knapsack_memoization(p, wt, m, n - 1, memo)
    else:
        memo[n][m] = max(
            p[n - 1] + knapsack_memoization(p, wt, m - wt[n - 1], n - 1, memo),
            knapsack_memoization(p, wt, m, n - 1, memo)
        )
    return memo[n][m]

def knapsack_memoization_wrapper(p, wt, m, n):
    memo = [[-1] * (m + 1) for _ in range(n + 1)]
    return knapsack_memoization(p, wt, m, n, memo)

def draw_grid(grid):
    print(tabulate(grid, tablefmt="grid"))

def sol(v, wt, m, n):
    i = n
    j = m
    ans = [0] * n
    objects = [i + 1 for i in range(n)]
    while i > 0 and j > 0:
        if v[i][j] == v[i - 1][j]:
            ans[i - 1] = 0
        else:
            ans[i - 1] = 1
            j -= wt[i - 1]
        i -= 1
    print("\n" + str(objects) + ": " + str(ans))

# Taking the input from the user
m = int(input("Enter the capacity of Knapsack: "))
n = int(input("Enter the number of objects: "))
profit = []
weight = []
for i in range(n):
    p = int(input("\nEnter the profit of object " + str(i + 1) + " : "))
    wt = int(input("Enter the weight of object " + str(i + 1) + " : "))
    profit.append(p)
    weight.append(wt)

# Calling the 0-1 Knapsack function
print("\nDynamic Programming Grid (Memoization):")
memo_grid = [[-1] * (m + 1) for _ in range(n + 1)]
knapsack_memoization(profit, weight, m, n, memo_grid)
draw_grid(memo_grid)
print("Maximum Profit (Memoization):", memo_grid[n][m])
```

```python
from tabulate import tabulate

def knapsack_bottom_up(p, wt, m, n):
    # Initializing the DP grid
    dp = [[0 for i in range(m + 1)] for j in range(n + 1)]

    # Filling the DP grid
    for i in range(1, n + 1):
        for w in range(1, m + 1):
            if wt[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], p[i - 1] + dp[i - 1][w - wt[i - 1]])
            else:
                dp[i][w] = dp[i - 1][w]

    # Finding the maximum profit
    max_profit = dp[n][m]

    # Displaying the Grid
    print("The Maximum Profit: " + str(max_profit))
    print("\nThe Grid")
    print(tabulate(dp, tablefmt="grid"))

    # Finding the solution
    i = n
    j = m
    ans = [0] * n
    objects = [i for i in range(1, n + 1)]
    while i > 0 and j > 0:
        if dp[i][j] == dp[i - 1][j]:
            ans[i - 1] = 0
        else:
            ans[i - 1] = 1
            j -= wt[i - 1]
        i -= 1
    print("\n" + str(objects) + ": " + str(ans))

# Taking the input from the user
m = int(input("Enter the capacity of Knapsack: "))
n = int(input("Enter the number of objects: "))
profit = []
weight = []
for i in range(n):
    p = int(input("\nEnter the profit of object " + str(i + 1) + " : "))
    wt = int(input("Enter the weight of object " + str(i + 1) + " : "))
    profit.append(p)
    weight.append(wt)

# Calling the 0-1 Knapsack function
knapsack_bottom_up(profit, weight, m, n)
```
